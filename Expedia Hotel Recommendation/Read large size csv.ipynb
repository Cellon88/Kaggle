# Read large size csv as batch processing

```python
import pandas as pd
import numpy as np
import tqdm
```

```python
from subprocess import check_output
print(check_output(["ls", "input"]).decode("utf8"))
```

```python
def process_frame(df):
    # process data frame
    return len(df)
```

## Read as FileReader

```python
if __name__ == '__main__':
    reader = pd.read_csv('input/train.csv', iterator=True, chunksize=100000)

    result = 0
    for df in reader:
        # process each data frame
        result += process_frame(df)
        print("Reading {} rows of data".format(result))
```
